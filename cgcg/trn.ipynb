```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import imutils
img1 = cv2.imread("cat.jpg")

rows, cols, _ = img1.shape

M = np.float32([[0, 1, 100], [1, 0, 50]])
t_img = cv2.warpAffine(img1, M, (cols, rows))

h_img = cv2.resize(img1, (0, 0), fx=0.5, fy=0.5)
w_img = cv2.resize(img1, (0, 0), fx=2, fy=2)

angle = 25
scale = 1.0

center = (cols // 2, rows // 2)

M = cv2.getRotationMatrix2D(center, angle, scale)

r_img = cv2.warpAffine(img1, M, (rows+200,cols+200))
r2_img = imutils.rotate_bound(img1, 25)
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(t_img, cv2.COLOR_BGR2RGB))
plt.title('Translated')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(h_img, cv2.COLOR_BGR2RGB))
plt.title('Half Size')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(w_img, cv2.COLOR_BGR2RGB))
plt.title('Double Size')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(r_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated')
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(r2_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated')
plt.show()
```
