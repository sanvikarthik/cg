```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img1=cv2.imread("cat.jpg")
img2=cv2.blur(img1,(55,55))
img_smooth=cv2.bilateralFilter(img1,15,150,150)

plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(img_smooth)
plt.show()
```
